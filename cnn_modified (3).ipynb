```python
import os
import sys
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
from tqdm.notebook import tqdm
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Dataset, Subset
from torchvision import datasets, transforms
from sklearn.metrics import classification_report, confusion_matrix


def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

from google.colab import drive
drive.mount('/content/drive', force_remount=False)
SAVE_DIR = '/content/drive/MyDrive/comp576final'
print(f'Google Drive mounted. Loading split from: {SAVE_DIR}')

def get_device():
    if torch.cuda.is_available():
        return torch.device("cuda")
    else:
        return torch.device("cpu")

DEVICE = get_device()
print(f"Using device: {DEVICE}")

IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 40
LR = 1e-4
VAL_RATIO = 0.1
TEST_RATIO = 0.1

path = kagglehub.dataset_download("emmarex/plantdisease")
print(f"Download path: {path}")

DATA_DIR = os.path.join(path, 'PlantVillage')
print(f"Dataset directory: {DATA_DIR}")

transform_train = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

transform_eval = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

class DatasetWrapper(Dataset):
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform

    def __getitem__(self, index):
        x, y = self.subset[index]
        if self.transform:
            x = self.transform(x)
        return x, y

    def __len__(self):
        return len(self.subset)

raw_dataset = datasets.ImageFolder(root=DATA_DIR)
classes = raw_dataset.classes
print(f"Found {len(classes)} classes. Total images: {len(raw_dataset)}")
total_size = len(raw_dataset)
split_path = os.path.join(SAVE_DIR, 'data_split_indices.pth')
if os.path.exists(split_path):
    print(f"Loading pre-saved data split indices from {split_path}...")
    split_indices = torch.load(split_path)
    if split_indices['total_size'] != total_size:
        print(f"Dataset size mismatch! Saved: {split_indices['total_size']}, Current: {total_size}")
        raise FileNotFoundError
    train_subset = Subset(raw_dataset, split_indices['train_indices'])
    val_subset = Subset(raw_dataset, split_indices['val_indices'])
    test_subset = Subset(raw_dataset, split_indices['test_indices'])
    print(f"Loaded saved split (seed={split_indices['random_seed']}):")
    print(f"   Train: {len(train_subset)} | Val: {len(val_subset)} | Test: {len(test_subset)}")

else:
    print("No pre-saved split found. Generating new split")
    val_size = int(VAL_RATIO * total_size)
    test_size = int(TEST_RATIO * total_size)
    train_size = total_size - val_size - test_size

    generator = torch.Generator().manual_seed(42)
    train_subset, val_subset, test_subset = random_split(
        raw_dataset, [train_size, val_size, test_size], generator=generator
    )
    split_indices = {
        'train_indices': train_subset.indices,
        'val_indices': val_subset.indices,
        'test_indices': test_subset.indices,
        'total_size': total_size,
        'random_seed': 42
    }
    torch.save(split_indices, split_path)
    print(f"Saved new split indices to {split_path}")
train_set = DatasetWrapper(train_subset, transform=transform_train)
val_set = DatasetWrapper(val_subset, transform=transform_eval)
test_set = DatasetWrapper(test_subset, transform=transform_eval)

NUM_WORKERS = 2

train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_set, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(test_set, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

class PlantCNN(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1), nn.BatchNorm2d(32), nn.ReLU(inplace=True), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), nn.BatchNorm2d(64), nn.ReLU(inplace=True), nn.MaxPool2d(2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True), nn.MaxPool2d(2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1), nn.BatchNorm2d(256), nn.ReLU(inplace=True), nn.MaxPool2d(2),
        )
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.classifier = nn.Sequential(
            nn.Linear(256, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.global_pool(x)
        x = x.view(x.size(0), -1)
        out = self.classifier(x)
        return out

class EarlyStopper:
    def __init__(self, patience=7, min_delta=0.001):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = float('inf')
        self.early_stop = False

    def __call__(self, val_loss):
        if val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
        elif val_loss > self.best_loss + self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        return self.early_stop

model = PlantCNN(num_classes=len(classes)).to(DEVICE)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LR)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3)
early_stopper = EarlyStopper(patience=8, min_delta=0.0001)
train_losses, val_accuracies = [], []
best_acc = 0.0
print(f"Starting training")
for epoch in range(EPOCHS):
    model.train()
    total_loss, correct, total = 0, 0, 0
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", leave=True)
    for imgs, labels in loop:
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
        loop.set_postfix(loss=loss.item(), acc=correct/total)
    avg_train_loss = total_loss / len(train_loader)
    train_losses.append(avg_train_loss)
    model.eval()
    val_correct, val_total, val_loss_sum = 0, 0, 0
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
            outputs = model(imgs)
            val_loss_sum += criterion(outputs, labels).item()
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    val_acc = val_correct / val_total
    avg_val_loss = val_loss_sum / len(val_loader)
    val_accuracies.append(val_acc)
    scheduler.step(avg_val_loss)
    current_lr = optimizer.param_groups[0]['lr']
    print(f"   Epoch {epoch+1} Summary: Train Loss={avg_train_loss:.4f} | Val Loss={avg_val_loss:.4f} | Val Acc={val_acc:.4f} | LR={current_lr:.1e}")
    if val_acc > best_acc:
        best_acc = val_acc
        torch.save(model.state_dict(), os.path.join(SAVE_DIR, "best_cnn_model.pth"))
    if early_stopper(avg_val_loss):
        print(f"Early stopping triggered at Epoch {epoch+1}")
        break
print("Training finished.")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(val_accuracies, label='Val Accuracy', color='orange')
plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.legend()
plt.tight_layout()
plt.show()

print("Evaluating on Test Set")
model.load_state_dict(torch.load(os.path.join(SAVE_DIR, "best_cnn_model.pth"), map_location=DEVICE))
model.eval()
y_true, y_pred = [], []
with torch.no_grad():
    for imgs, labels in tqdm(test_loader, desc="Testing"):
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        outputs = model(imgs)
        _, preds = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

print("Classification Report:")
try:
    print(classification_report(y_true, y_pred, target_names=classes))
except Exception:
    print(classification_report(y_true, y_pred))

if len(classes) < 50:
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=False, fmt='d', cmap='Greens', xticklabels=classes, yticklabels=classes)
    plt.title("Confusion Matrix")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.xticks(rotation=90)
    plt.show()

def imshow(img, title):
    img = img.cpu().numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = std * img + mean
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    plt.title(title, fontsize=8)
    plt.axis('off')

dataiter = iter(test_loader)
images, labels = next(dataiter)
num_show = min(5, len(images))
images = images[:num_show]
labels = labels[:num_show]
model.eval()
with torch.no_grad():
    outputs = model(images.to(DEVICE))
    _, preds = torch.max(outputs, 1)
plt.figure(figsize=(15, 4))
print("Sample Predictions:")
for i in range(num_show):
    plt.subplot(1, num_show, i+1)
    pred_label = classes[preds[i]]
    true_label = classes[labels[i]]
    color = "green" if pred_label == true_label else "red"
    title = f"T: {true_label}\nP: {pred_label}"
    imshow(images[i], title)
plt.show()
```

```python
test_accuracy = 100 * sum([1 for true, pred in zip(y_true, y_pred) if true == pred]) / len(y_true)
print(f"\nTest Accuracy: {test_accuracy:.2f}%")
```
